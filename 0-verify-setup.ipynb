```python
# 3. Check environment variables
import os
from dotenv import load_dotenv
load_dotenv()

endpoint = os.getenv("AZURE_AI_PROJECT_ENDPOINT")
model = os.getenv("AZURE_AI_MODEL_DEPLOYMENT_NAME", "gpt-4o")

if endpoint:
    print(f"✅ Project endpoint: {endpoint[:50]}...")
    print(f"✅ Model deployment: {model}")
else:
    print("❌ AZURE_AI_PROJECT_ENDPOINT not set!")
    print("   Copy .env.example to .env and fill in your values.")
```

```python
# 4. Check Azure authentication
import asyncio
from azure.identity.aio import AzureCliCredential

async def check_auth():
    async with AzureCliCredential() as credential:
        token = await credential.get_token("https://management.azure.com/.default")
        print(f"✅ Azure CLI authenticated (token expires: {token.expires_on})")

await check_auth()
```

```python
# 5. Quick test: can we create a client and get a response?
from agent_framework import ChatAgent
from agent_framework.azure import AzureAIAgentClient
from azure.identity.aio import AzureCliCredential

async def quick_test():
    async with (
        AzureCliCredential() as credential,
        AzureAIAgentClient(async_credential=credential) as client,
    ):
        agent = client.create_agent(
            name="TestAgent",
            instructions="Reply with exactly: 'Setup complete!'",
        )
        result = await agent.run("Hello")
        print(f"✅ Agent response: {result.text}")
        print("\n🎉 ALL CHECKS PASSED — You're ready for the workshop!")

await quick_test()
```
